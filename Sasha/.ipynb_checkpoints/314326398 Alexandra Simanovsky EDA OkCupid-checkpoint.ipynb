## Graph No. 2 -  If job hav e any influence of bady structure- compare to f\m

**We need make nec col with less body diffinitions **


```python
df['body_type'].unique()
```

```python

def bod(x):
    if x== 'a little extra' or  x== 'curvy' or  x== 'average' or x=='overweight':
        return 'Overweight'
    else:
        return 'Not Overweight'
df['newBody']= df['body_type'].dropna().apply(lambda x:bod(x))   

df.head()
```

```python
sns.boxplot(data=df,x='newBody',y='age',hue='sex',palette=dict(m='g',f='r'))

plt.legend(loc='upper center', bbox_to_anchor=(1.05,1))
plt.title('Age vs body type in F\M',fontsize=20)
plt.show()
```

1. The plot look almost same.
2. The females up to 40 defined them as overweight and not overweight, the age distribution in this categoris  among females is  that the 1/4 in the now overweight is in more adult age in compereson to the overweight group.

We can lear so mach from this plot, the bar plot give us a lot of information.

```python
# Set up the matplotlib figure
f, axes = plt.subplots(1, 2, figsize=(7, 7), sharex=True)
sns.despine(left=True)

ax=sns.boxplot(data=df,x='newBody',y='age',hue='sex',palette=dict(m='g',f='r'), ax=axes[0])
ax.legend_.remove()

sns.violinplot(data=df,x='newBody',y='age',hue='sex',palette=dict(m='g',f='r'), ax=axes[1]);

plt.legend([]) # remvoe the vaiolin legend
plt.tight_layout()

plt.suptitle('Body Type  vs Age of Female or Male ', size=16)

 
plt.show()
```

```python
# Set up the matplotlib figure
f, axes = plt.subplots(2, 2, figsize=(7, 7), sharex=True)
sns.despine(left=True)


# Plot a simple histogram with binsize determined automatically

sns.distplot(df['age'], kde=False, color="b", ax=axes[0, 0])


# Plot a kernel density estimate and rug plot
sns.distplot(df['age'], hist=False, rug=True, color="r", ax=axes[0, 1])

# Plot a filled kernel density estimate
sns.distplot(df['age'], hist=False, color="g", kde_kws={"shade": True}, ax=axes[1, 0])

# Plot a historgram and kernel density estimate
sns.distplot(df['age'], color="m", ax=axes[1, 1])

plt.setp(axes, yticks=[])
plt.suptitle('Age distribution Fig ', size=16)
plt.tight_layout()
```
